## DevOps

[[back to top](#Sections)]

<center> Atlassian. (n.d.). What is DevOps? Retrieved June 9, 2018, from https://www.atlassian.com/devops</center>
<hr>
The Atlassian, article generated by the company explain well about DevOps. Many organizations are using agile methodologies from developing software and planning project management but there are still existing some issue to get the codes for the system out door. The article recommend that your organization has with the issues then try the DevOps way. The DevOps focuses on culture and collaboration to break the barrier between developments and operations.Thus, it focuses on trust each other that is most important factor to success. “Building a culture of shared responsibility, transparency and faster feedback is the foundation of every high performing DevOps team.” In addition, the DevOps emphasis automation to eliminate repetitive manual work by people. People might cause a mistake in the repetitive works so it need to make into an automating job to avoid from a mistake. Also, the DevOps is able to measure the qualitive jobs in common ways. Below are the list:
<hr>

- How long did it take to go from development to deployment?
- How often do recurring bugs or failures happen?
- How long does it take to recover after a system failure?
- How many people are using your product right now?
- How many users did you gain / lose this week? [7]

## Project Management Methodologies Comparison

[[back to top](#Sections)]

<center>Scheid, J., & McDonough, M. (2015, July 22). Project Management Methodologies Comparison: Pros and Cons of Various PM Methods. Retrieved June 23, 2018, from https://www.brighthubpm.com/methods-strategies/67087-project-managementmethodologieshow-do-they-compare/</center>
<hr>
The author, article introduce about methodologies for IT project and project management. Methodology is a theory and idea to explain a situation. There are several types of methodologies pervasive in the market. Agile, Waterfall, Risk, Change, Quality PRINCE2 and Six Sigma, those are the methodologies to manage project to achieve goal and create value. The author explains
each methodology and some methodology comparing each other. 
<hr>

- [[Agile](https://www.atlassian.com/agile)] uses the best process through “teams, customer involvement, and the ability to analyze and quickly control changes to the project scope at inception and throughout the lifecycle of the project."
- [[Waterfall](https://en.wikipedia.org/wiki/Waterfall_model)] is traditional methodology that assign a project with clearly set goals and timelines and it is not allowing to change scope.
- [[Change management](https://changeactivation.com/change-management-models/)] “helps a company prepare for planned and unplanned changes. This change can be forced upon an organization from both internal or external factors.”
- [[Risk management](https://definedterm.com/risk_management_methodology)] is “described as identifying risks, measuring their potential for harm, and creating plans to deal with the threats.”
- [[Quality management](https://en.wikipedia.org/wiki/Quality_management)] has two goals “ensuring a quality end-product and ensuring that all of the process involved during the project lifecycle are carried out efficiently.”
- [[PRINCE2](https://en.wikipedia.org/wiki/PRINCE2)] developed in the United Kingdom includes six variables, “costs, timescales, quality, scope, risk, and benefits.”
- [[Six Sigma](https://en.wikipedia.org/wiki/Six_Sigma)] defines “statistical concept that measures a process in terms of defects.” Agile developed through Six Sigma.
